```python
import os
import cv2
import numpy as np

dataset_path = "digits"

I = []
J = []

total_images = 0
loaded_images = 0
failed_images = 0

print("Starting dataset loading...\n")

for label in range(10):  # 0 to 9
    folder = os.path.join(dataset_path, str(label))

    # Check if folder exists
    if not os.path.exists(folder):
        print(f"âš ï¸ Folder missing: {folder}")
        continue

    images = os.listdir(folder)
    total_images += len(images)

    for img_name in images:
        img_path = os.path.join(folder, img_name)

        # Read image as grayscale
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

        # Error check
        if img is None:
            print(f"âŒ Failed to load image: {img_path}")
            failed_images += 1
            continue

        # Resize to 28x28
        img = cv2.resize(img, (28, 28))

        # Normalize (0â€“255 â†’ 0â€“1)
        img = img.astype(np.float32) / 255.0

        I.append(img)
        J.append(label)
        loaded_images += 1

        # Progress update every 50 images
        if loaded_images % 50 == 0:
            print(f"âœ… Loaded {loaded_images}/{total_images} images")

print("\nğŸ“Š Dataset loading summary:")
print("Total images found :", total_images)
print("Successfully loaded:", loaded_images)
print("Failed images      :", failed_images)

# Convert to numpy arrays
I = np.asarray(I, dtype=np.float32)
J = np.asarray(J, dtype=np.int64)

print("\nFinal shapes:")
print("X shape:", I.shape)  # (num_samples, 28, 28)
print("y shape:", J.shape)  # (num_samples,)
```
